```python
from main import parse
import fifo, lru, optff


files = ['../tests/test1.txt', '../tests/test3.txt', '../tests/test2.txt']

index = 1

print("File, k, m, FIFO, LRU, OPTFF")

for file in files:
    k, m, r = parse(file)
    fifo_misses = fifo.fifo(k, r)
    lru_misses = lru.LRU_Test(k, r)
    optff_misses = optff.OPT_Test(k, r)
    

    print(index, k , m,  fifo_misses, lru_misses, optff_misses)
    index+=1
```

Does OPTFF have the fewest misses?
Yes,
OPTFF proves to be the most effective algorithm of the three tested. This is because it looks ahead on the input data and makes a perfect eviction choice.

How does FIFO compare to LRU?

LRU generally does at least equal or fewer misses to the FIFO algorithm. This is because FIFO just evicts the oldest item, rather than the one being the least frequenltly used - LRU is being smarter about which ID to evict.
